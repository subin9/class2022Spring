```python
!pip install gradio
import gradio as gr
import requests
import pandas as pd

def fire_status(sidoname):

    URL = ('http://apis.data.go.kr/1661000/EmergencyStatisticsService/getEmgDispatchResultStats?serviceKey=EPb7a19mqphUeFDHld9SJ7KccAkM4uQBnlKNJ8%2FpkNBCSTF5E%2BjT6nt0CgI25moei7n9%2BIrSurGjEbJArkYnlQ%3D%3D&numOfRows=100&pageNo=1&resultType=json&sidoHqOgidNm='+sidoname+'&rcptYm=202202&gutTyCdNm=정상')

    response = requests.get(URL)
    r=response.json()

    items = r['body']['items']
    Regname=[item['rsacGutFsttOgidNm']for item in items]
    Emenum=[item['trnfCo']for item in items]
    Patnum=[item['trnfPcnt']for item in items]

    Regname = pd.Series(Regname)
    Emenum = pd.Series(Emenum)
    Patnum = pd.Series(Patnum)
    df=pd.DataFrame({'소방서 이름':Regname, '이송 횟수':Emenum, '이송 환자 수':Patnum})
    
    df2 = {'소방서 이름': '계', '이송 횟수':sum([item['trnfCo']for item in items]), '이송 환자 수': sum([item['trnfPcnt']for item in items])}
    df = df.append(df2, ignore_index = True) #실용적이려면 합계가 꼭 필요하다.추가해주자.
    df.index+=1 # index가 0부터 시작하면 보기 싫다.
    
    return df

iface = gr.Interface(fire_status, gr.inputs.Dropdown(['소방재난본부', '부산소방재난본부', '대구소방안전본부', '인천소방본부', '광주소방본부', '대전소방본부', '울산소방본부', '경기도소방재난본부', '강원소방본부', '충북소방본부', '충청남도소방본부', '전북소방본부', '전남소방본부', '경북소방본부', '경남소방본부', '제주소방안전본부', '세종소방본부']), outputs=['dataframe'])
iface.launch(debug=True)


```
